## Overview 
In this tutorial, we will cover the basics of constructing dataset and iterators so that we can train models using gradient descent. 

The best tutorial can be found in the official website (https://pytorch.org/tutorials/beginner/data_loading_tutorial.html).

```python
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class MyStupidDataset(Dataset):#继承Dataset类
    def __init__(self):
        super(MyStupidDataset, self).__init__()
        self.data = torch.randn([1024, 10, 10])#1024个10*10的Tensor
    def __len__(self):
        return 1024
    def __getitem__(self, idx):
        return self.data[idx, :, :] 
```

```python
my_stupid_dataset = MyStupidDataset()
```

```python
my_data_loader = DataLoader(my_stupid_dataset, batch_size=64, shuffle=True)#my_stupid_dataset从哪里导入？batch_size抽几个数？shuffle是否打乱？
```

```python
for i in my_data_loader:
    print(i)
```

```python
# A common Pattern
class MyDictDataset(Dataset):
    def __init__(self):
        super(MyDictDataset, self).__init__()
        self.x = torch.randn(1024, 10)
        self.y = torch.randn(1024)
    def __len__(self):
        return 1024
    def __getitem__(self, idx):
        return {'x':self.x[idx,:],'y':self.y[idx]}
```

```python
my_dict_dataset = MyDictDataset()
my_data_loader = DataLoader(my_dict_dataset, batch_size=64, shuffle=True)
for batch in my_data_loader:
    print(batch['x'])
    print(batch['y'])
```

为什么要写自己的Dataset?
    文本非常大，我们可能要做一个iterator，有几种方法来实现：①直接打开该文件，读取其中的一部分文本，直接用IO的方式来写，这样保证了我们读取数据所占用的内存或显存较小，但读写速度较慢②全部读到内存，在内存中做iteration③每次做iteration时，重新分词，然后对其对应的token进行更新，通过写自己的Dataset的方式，可以把操作在不同阶段进行处理。例如，每次重新读取文本比较慢，可以先将其对应的Embedding先找出来，自己构建一个数据集，但这样做的问题在于占用大量内存和显存，读入速度慢，但可以节约找embedding的layer的时间
